데이터 프레임의 메서드를 사용하여 열을 삭제, 판다스에서는 데이터프레임의 행이나 열을 삭제하는 drop() 매서드를 제공, drop() 메서드로 열을 삭제하려면 첫번째 매개변수에 삭제하려는 열 이름을 전달하고 axis매개변수를 1로 지정!

```python
ns_book = ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

```python
ns_book = ns_df.drop(['부가기호','Unnamed: 13'], axis=1)
#여러 개의 열을 삭제하려면 리스트 형식 사용합니다.
ns_book.head()
```

지금까지는 ns_df데이터프레임에서 열을 삭제한 결과를 ns_book이라는 변수에 저장했습니다. drop() 매서드에 inplace매개변수를 True로 지정하면, 현재 선택한 데이터프레임을 바로 수정할 수도 있습니다. 앞서 생성한 ns_book데이터프레임의 '주제분류번호'열을 삭제해보자.

```python
ns_book.drop('주제분류번호', axis=1, inplace=True)
#복사본이 아니라, 원본(ns_book) 그 자체를 수정해 버립니다.
#책에서 이 옵션을 쓴 이유는, 앞으로 계속 '주제분류번호'가 없는 상태의 ns_book을 쓸 것이기 때문에, 번거롭게 다시 저장하지 않고 그 자리에서 바로 없애버린 거예요.
ns_book.head()
```

판다스는 비어있는 값을 NaN으로 표시합니다. drop()메서드와 비슷한 dropna()메서드는 기본적으로 NaN이 하나 이상 포함된 행이나 열을 삭제. axis=1로 지정하여 ns_df데이터프레임에서 NaN이 포함된 열을 삭제! 14개의 열이5개로 확 줄어듭니다. 삭제된열에는 NaN이 적어도 하나 이상 있었나 봅니다.

```python
ns_book = ns_df.dropna(axis=1)
ns_book.head()
```

```python
ns_book = ns_df.dropna(axis=1, how='all')
# 모든 값이 NaN인 열을 삭제하려면, how='all'
ns_book.head()
```

지금까지 열 삭제하는 방법, loc메서드로 슬라이스 연산자나 불리언 배열을 사용할 수 있다. drop()메서드, NaN이 들어있는열을 삭제하려면 dropna()메서드에 axis=1을 지정한다.

# 중복된 행 찾기

판다스 데이터프레임의 중복된 행은 duplicated()메서드를 사용

중복된 행 중에서 처음 행을 제외한 나머지 행은 True로, 그외에 중복되지 않은 나머지 모든 행은 False로 표시한 불리언 배열을 반환한다.

```python
sum(ns_book.duplicated())
# 중복된 행이 없다는 뜻!! 이처럼 duplicated()메서드는 기본적으로 데이터프레임에 있는 모든 열을 기준으로 중복된 행을 찾는다!
```

```python
sum(ns_book.duplicated(subset=['도서명','저자','ISBN']))
```

'도서명','저자','ISBN'을 기준으로 중복된 행이 있는지를 찾아보면 어떨까요?
일부 열을 기준으로 중복된 행을 찾으려면 duplicated()메서드의 **subset매개변수**에 기준열을 나열한다!!

'도서명','저자','ISBN'을 기준으로 찾으면 중복된 행이 2만개가 넘는다!!

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'], keep=False) #중복된 행을 모두 True로 표시한 불리언 배열을 반환합니다.
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

keep='first' (기본값): "첫 번째 놈(A)은 남겨두고, 나머지(B, C)만 중복이라고 표시해." (A는 False, B와 C는 True)

keep='last': "마지막 놈(C)만 남겨두고, 앞의 놈들(A, B)을 중복이라고 표시해." (C는 False, A와 B는 True)

keep=False: "누가 처음인지 마지막인지 안 따질 거야. 중복된 게 하나라도 있으면 셋 다(A, B, C) 중복이라고 표시해!" (A, B, C 모두 True)

# 그룹별로 모으기

대출건수 열이 중요하다. 같은 도서의 대출건수를하나로 합치기. ex.'혼공 데이터 분석' 도서가 두 권이라면 하나로 합치려고 한다. groupby()메서드를 사용.


```python
count_df = ns_book[['도서명','저자','ISBN','권','대출건수']] #그룹으로 묶을 기준 열
print(count_df)
```

```python
count_df = ns_book[['도서명','저자','ISBN','권','대출건수']] #그룹으로 묶을 기준 열
#필요한 재료만 도마 위에 올리기
#**"합칠 때 기준이 되는 정보"**와 **"더해야 할 대출건수"**만 쏙 뽑아서 가벼운 가방(count_df)을 먼저 만드는 거
print(count_df)
```

```python
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False)
#끼리끼리 모으기
#안에 적힌 네 가지 정보(제목, 저자, ISBN, 권)가 모두 똑같은 책들만 골라 하나의 그룹으로 묶어줍니다
loan_count = group_df.sum()
#팀별로 점수 합산하기
#그룹 안에 있는 책들의 대출건수를 다 더해서 하나의 숫자로 만들어줍니다.
print(loan_count)
```

```python
loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum()
#"지름길 이용하기"
#"그룹으로 묶고(.groupby), 바로 합쳐(.sum)!"라고 한 번에 시키는 거죠.
loan_count.head()
```
